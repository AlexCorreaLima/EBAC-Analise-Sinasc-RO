#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd

# Carregar o CSV
df = pd.read_csv('SINASC_RO_2019.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
import matplotlib.pyplot as plt

painel_vars = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.flatten()

for i, var in enumerate(painel_vars):
    df[var].value_counts().sort_index().plot(kind='bar', ax=axs[i], color='lightcoral')
    axs[i].set_title(f'Frequência da variável {var}')
    axs[i].set_xlabel(var)
    axs[i].set_ylabel('Frequência')
    axs[i].grid(axis='y')

plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
import matplotlib.pyplot as plt

# Agrupamento por 'CONSULTAS' e cálculo da média do APGAR5
media_apgar5_por_consultas = df.groupby('CONSULTAS')['APGAR5'].mean()

# Plotando o gráfico de barras
media_apgar5_por_consultas.plot(kind='bar', color='skyblue', figsize=(10,6))

plt.title('Média do APGAR5 por número de Consultas')
plt.xlabel('Consultas Pré-natal')
plt.ylabel('Média do APGAR5')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
import matplotlib.pyplot as plt

# Lista com as variáveis a serem analisadas
variaveis_painel = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']

# Criando o painel de gráficos
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.flatten()

for i, var in enumerate(variaveis_painel):
    df[var].value_counts().sort_index().plot(kind='bar', ax=axs[i], color='lightcoral')
    axs[i].set_title(f'Frequência da variável {var}')
    axs[i].set_xlabel(var)
    axs[i].set_ylabel('Frequência')
    axs[i].grid(axis='y')

plt.tight_layout()
plt.show()
```
